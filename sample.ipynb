```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.scatterplot(iris, 
                x = "petal_length",
                y = "petal_width",
                hue = "species",
                size = "sepal_length",
                )
plt.title("Iris Petal Length vs Petal Width")
plt.ylabel("Petal Witdh [cm]")
plt.xlabel("Petal Length [cm]")
plt.show()
```

There is a strong linear trend between petal length and width. With fairly decent separation between species.
